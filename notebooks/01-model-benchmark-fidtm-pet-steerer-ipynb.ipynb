```python
import json
from pathlib import Path

import numpy as np
import pandas as pd
from scipy.io import loadmat

# -----------------------------
# Mall ground truth loader
# -----------------------------
def load_mall_samples(frames_dir: Path, gt_mat_path: Path) -> pd.DataFrame:
    """
    Load Mall dataset frame paths and ground-truth counts.

    Mall frames are named seq_000001.jpg, seq_000002.jpg, ...
    mall_gt.mat usually contains a 'count' array.
    """
    mat = loadmat(gt_mat_path)
    print("Mall .mat keys:", [k for k in mat.keys() if not k.startswith("__")])

    if "count" not in mat:
        raise KeyError("Could not find 'count' key in mall_gt.mat")

    counts = np.asarray(mat["count"]).reshape(-1)
    frame_paths = sorted(frames_dir.glob("*.jpg"))

    n = min(len(frame_paths), len(counts))

    rows = []
    for i in range(n):
        rows.append({
            "dataset": "Mall",
            "frame_id": frame_paths[i].stem,
            "image_path": str(frame_paths[i]),
            "gt_count": int(counts[i]),
        })

    return pd.DataFrame(rows)


# -----------------------------
# FDST ground truth loader
# -----------------------------
def count_points_in_fdst_json(json_path: Path) -> int:
    """
    Count annotated head points in one FDST JSON file.
    Handles common JSON structures safely.
    """
    with open(json_path, "r") as f:
        data = json.load(f)

    # Common FDST format: list of point dicts or dict containing points
    if isinstance(data, list):
        return len(data)

    if isinstance(data, dict):
        for key in ["points", "annotations", "annPoints", "gt_points", "locations"]:
            if key in data and isinstance(data[key], list):
                return len(data[key])

        # fallback: count values that look like point lists
        for value in data.values():
            if isinstance(value, list) and len(value) > 0:
                first = value[0]
                if isinstance(first, (list, tuple, dict)):
                    return len(value)

    raise ValueError(f"Unknown FDST JSON format: {json_path}")


def load_fdst_samples(images_dir: Path, jsons_dir: Path, max_frames: int = 6000) -> pd.DataFrame:
    """
    Load first max_frames FDST frames and ground-truth counts from matching JSON files.
    """
    image_paths = sorted(images_dir.glob("*.jpg"))[:max_frames]

    rows = []
    missing_json = 0

    for img_path in image_paths:
        json_path = jsons_dir / f"{img_path.stem}.json"

        if not json_path.exists():
            missing_json += 1
            continue

        gt_count = count_points_in_fdst_json(json_path)

        rows.append({
            "dataset": "FDST",
            "frame_id": img_path.stem,
            "image_path": str(img_path),
            "gt_count": int(gt_count),
        })

    if missing_json:
        print("Missing FDST JSON files:", missing_json)

    return pd.DataFrame(rows)


mall_df = load_mall_samples(MALL_FRAMES_DIR, MALL_GT_MAT)
fdst_df = load_fdst_samples(FDST_IMAGES_DIR, FDST_JSONS_DIR, MAX_FDST_FRAMES)

print("\nMall samples:", len(mall_df))
print(mall_df.head())
print(mall_df["gt_count"].describe())

print("\nFDST samples:", len(fdst_df))
print(fdst_df.head())
print(fdst_df["gt_count"].describe())

all_samples_df = pd.concat([mall_df, fdst_df], ignore_index=True)

samples_csv = BENCH_DIR / "benchmark_samples_mall_fdst_6000.csv"
all_samples_df.to_csv(samples_csv, index=False)

print("\nSaved samples CSV:", samples_csv)
print("Total samples:", len(all_samples_df))
```

```python
from pathlib import Path

def show_tree_limited(root: Path, max_depth: int = 3, max_items: int = 40):
    print(f"\nROOT: {root}")
    if not root.exists():
        print("Missing")
        return

    def _walk(path: Path, depth: int, prefix: str):
        if depth > max_depth:
            return

        items = sorted(list(path.iterdir()), key=lambda x: (x.is_file(), x.name.lower()))
        shown = items[:max_items]

        for item in shown:
            if item.is_dir():
                print(prefix + f"[DIR]  {item.name}")
                _walk(item, depth + 1, prefix + "  ")
            else:
                print(prefix + f"[FILE] {item.name}")

        if len(items) > max_items:
            print(prefix + f"... {len(items) - max_items} more items not shown")

    _walk(root, 1, "")


for repo_name in ["FIDTM", "PET", "CLTR", "STEERER"]:
    show_tree_limited(REPOS_DIR / repo_name, max_depth=3, max_items=30)
```

```python
import cv2
import time
import numpy as np
import torch
import torch.nn.functional as F
from PIL import Image
from torchvision import transforms

FIDTM_MEAN = [0.485, 0.456, 0.406]
FIDTM_STD = [0.229, 0.224, 0.225]

fidtm_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=FIDTM_MEAN, std=FIDTM_STD),
])

def resize_to_multiple_of_16(image_rgb: np.ndarray):
    h, w = image_rgb.shape[:2]
    new_h = h - (h % 16)
    new_w = w - (w % 16)

    if new_h <= 0 or new_w <= 0:
        raise ValueError(f"Invalid resized shape from {image_rgb.shape}")

    resized = cv2.resize(image_rgb, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    return resized, (h, w), (new_h, new_w)


@torch.no_grad()
def predict_fidtm_count(image_path: str):
    """
    Returns:
        pred_count: float
        inference_time: seconds
    """
    bgr = cv2.imread(image_path)
    if bgr is None:
        raise FileNotFoundError(image_path)

    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    rgb_resized, original_shape, resized_shape = resize_to_multiple_of_16(rgb)

    x = fidtm_transform(Image.fromarray(rgb_resized)).unsqueeze(0).to(DEVICE)

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    start = time.perf_counter()
    output = fidtm_model(x)

    if isinstance(output, (list, tuple)):
        output = output[0]

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    elapsed = time.perf_counter() - start

    pred_count = float(output.detach().sum().item())

    return pred_count, elapsed


sample_row = mall_df.iloc[0]
pred, sec = predict_fidtm_count(sample_row["image_path"])

print("Sample:", sample_row["image_path"])
print("GT count:", sample_row["gt_count"])
print("FIDTM pred:", pred)
print("Inference time:", sec)
print("FPS:", 1.0 / sec if sec > 0 else None)
```

```python
import cv2
import numpy as np
import torch
from PIL import Image

def extract_points_from_fidtm_map(
    pred_map: np.ndarray,
    threshold: float = 0.10,
    nms_kernel: int = 15,
):
    """
    Convert FIDTM response map into predicted head points using local maxima.

    Returns:
        List of (x, y, score)
    """
    pred_map = np.squeeze(pred_map)

    if pred_map.ndim != 2:
        raise ValueError(f"Expected 2D map, got shape {pred_map.shape}")

    # Remove negative values
    pred_map = np.maximum(pred_map, 0)

    # Local max filter
    kernel = np.ones((nms_kernel, nms_kernel), np.uint8)
    local_max = cv2.dilate(pred_map, kernel)

    peaks = (pred_map == local_max) & (pred_map >= threshold)

    ys, xs = np.where(peaks)

    points = []
    for x, y in zip(xs, ys):
        points.append((float(x), float(y), float(pred_map[y, x])))

    return points


@torch.no_grad()
def predict_fidtm_points(image_path: str, threshold: float = 0.10, nms_kernel: int = 15):
    """
    Predict FIDTM head points and count.
    """
    bgr = cv2.imread(image_path)
    if bgr is None:
        raise FileNotFoundError(image_path)

    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    rgb_resized, original_shape, resized_shape = resize_to_multiple_of_16(rgb)

    x = fidtm_transform(Image.fromarray(rgb_resized)).unsqueeze(0).to(DEVICE)

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    start = time.perf_counter()
    output = fidtm_model(x)

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    elapsed = time.perf_counter() - start

    pred_map = output.squeeze().detach().cpu().numpy()

    points = extract_points_from_fidtm_map(
        pred_map,
        threshold=threshold,
        nms_kernel=nms_kernel,
    )

    return {
        "count": len(points),
        "points": points,
        "time_sec": elapsed,
        "map_shape": pred_map.shape,
    }


sample_row = mall_df.iloc[0]

for th in [0.03, 0.05, 0.07, 0.10, 0.15, 0.20]:
    result = predict_fidtm_points(sample_row["image_path"], threshold=th, nms_kernel=15)
    print(f"threshold={th:.2f} | pred_count={result['count']} | gt={sample_row['gt_count']} | time={result['time_sec']:.4f}s")
```

```python
from tqdm.auto import tqdm
import numpy as np
import pandas as pd

FIDTM_THRESHOLD = 0.20
FIDTM_NMS_KERNEL = 15

def compute_count_metrics(gt_counts, pred_counts):
    gt = np.array(gt_counts, dtype=np.float32)
    pred = np.array(pred_counts, dtype=np.float32)

    errors = np.abs(pred - gt)
    sq_errors = (pred - gt) ** 2

    mae = float(np.mean(errors))
    rmse = float(np.sqrt(np.mean(sq_errors)))

    return mae, rmse


def run_fidtm_count_benchmark(samples_df: pd.DataFrame, max_samples: int = None):
    rows = []

    if max_samples is not None:
        samples_df = samples_df.head(max_samples).copy()

    for _, row in tqdm(samples_df.iterrows(), total=len(samples_df)):
        result = predict_fidtm_points(
            row["image_path"],
            threshold=FIDTM_THRESHOLD,
            nms_kernel=FIDTM_NMS_KERNEL,
        )

        rows.append({
            "model": "FIDTM",
            "dataset": row["dataset"],
            "frame_id": row["frame_id"],
            "image_path": row["image_path"],
            "gt_count": int(row["gt_count"]),
            "pred_count": int(result["count"]),
            "abs_error": abs(int(result["count"]) - int(row["gt_count"])),
            "sq_error": (int(result["count"]) - int(row["gt_count"])) ** 2,
            "time_sec": float(result["time_sec"]),
            "fps": float(1.0 / result["time_sec"]) if result["time_sec"] > 0 else 0.0,
            "num_pred_points": len(result["points"]),
        })

    return pd.DataFrame(rows)


quick_samples = pd.concat([
    mall_df.head(100),
    fdst_df.head(100),
], ignore_index=True)

fidtm_quick_df = run_fidtm_count_benchmark(quick_samples)

display(fidtm_quick_df.head())

quick_summary = (
    fidtm_quick_df
    .groupby(["model", "dataset"])
    .agg(
        frames=("frame_id", "count"),
        mae=("abs_error", "mean"),
        rmse=("sq_error", lambda x: float(np.sqrt(np.mean(x)))),
        avg_fps=("fps", "mean"),
        avg_time_sec=("time_sec", "mean"),
        mean_gt=("gt_count", "mean"),
        mean_pred=("pred_count", "mean"),
    )
    .reset_index()
)

display(quick_summary)
```

```python
import json
import numpy as np
from scipy.io import loadmat

def load_mall_gt_points_map(gt_mat_path: Path):
    """
    Load Mall ground-truth point locations from mall_gt.mat.

    Returns:
        dict: frame_id -> list[(x, y)]
    """
    mat = loadmat(gt_mat_path)
    frame_data = mat["frame"]

    gt_map = {}

    for i in range(frame_data.shape[1]):
        frame_id = f"seq_{i+1:06d}"

        entry = frame_data[0, i]

        # Mall stores location points inside nested MATLAB structures.
        points = None

        def find_points(obj):
            if isinstance(obj, np.ndarray):
                if obj.ndim == 2 and obj.shape[1] == 2 and np.issubdtype(obj.dtype, np.number):
                    return obj
                for item in obj.flat:
                    found = find_points(item)
                    if found is not None:
                        return found
            return None

        points = find_points(entry)

        if points is None:
            gt_map[frame_id] = []
        else:
            gt_map[frame_id] = [(float(x), float(y)) for x, y in points]

    return gt_map


def load_fdst_points_from_json(json_path: Path):
    """
    Load FDST GT point locations from JSON.
    Returns list[(x, y)].
    """
    with open(json_path, "r") as f:
        data = json.load(f)

    points = []

    if isinstance(data, list):
        raw_points = data
    elif isinstance(data, dict):
        raw_points = None
        for key in ["points", "annotations", "annPoints", "gt_points", "locations"]:
            if key in data and isinstance(data[key], list):
                raw_points = data[key]
                break
        if raw_points is None:
            for value in data.values():
                if isinstance(value, list):
                    raw_points = value
                    break
    else:
        raw_points = []

    if raw_points is None:
        raw_points = []

    for p in raw_points:
        if isinstance(p, dict):
            if "x" in p and "y" in p:
                points.append((float(p["x"]), float(p["y"])))
            elif "point" in p and len(p["point"]) >= 2:
                points.append((float(p["point"][0]), float(p["point"][1])))
        elif isinstance(p, (list, tuple)) and len(p) >= 2:
            points.append((float(p[0]), float(p[1])))

    return points


def load_fdst_gt_points_map(jsons_dir: Path, frame_ids):
    gt_map = {}

    for frame_id in frame_ids:
        json_path = jsons_dir / f"{frame_id}.json"
        gt_map[frame_id] = load_fdst_points_from_json(json_path)

    return gt_map


mall_gt_points_map = load_mall_gt_points_map(MALL_GT_MAT)
fdst_gt_points_map = load_fdst_gt_points_map(FDST_JSONS_DIR, fdst_df["frame_id"].head(MAX_FDST_FRAMES).tolist())

print("Mall point frames:", len(mall_gt_points_map))
print("FDST point frames:", len(fdst_gt_points_map))

for fid in ["seq_000001", "seq_000002"]:
    print(fid, "Mall points:", len(mall_gt_points_map[fid]), mall_gt_points_map[fid][:3])

for fid in ["100_001", "100_002"]:
    print(fid, "FDST points:", len(fdst_gt_points_map[fid]), fdst_gt_points_map[fid][:3])
```

```python
from scipy.optimize import linear_sum_assignment
import numpy as np

def compute_localization_metrics(gt_points, pred_points, distance_threshold: float = 16.0):
    """
    Compute point-level localization Precision, Recall, and F1.

    A prediction is a true positive if it is matched to one GT point
    within distance_threshold pixels. Matching is one-to-one using Hungarian assignment.
    """
    gt = np.array(gt_points, dtype=np.float32)

    # pred_points can be [(x,y)] or [(x,y,score)]
    pred = np.array([(p[0], p[1]) for p in pred_points], dtype=np.float32)

    if len(gt) == 0 and len(pred) == 0:
        return {"tp": 0, "fp": 0, "fn": 0, "precision": 1.0, "recall": 1.0, "f1": 1.0}

    if len(gt) == 0:
        return {"tp": 0, "fp": len(pred), "fn": 0, "precision": 0.0, "recall": 1.0, "f1": 0.0}

    if len(pred) == 0:
        return {"tp": 0, "fp": 0, "fn": len(gt), "precision": 1.0, "recall": 0.0, "f1": 0.0}

    dists = np.linalg.norm(gt[:, None, :] - pred[None, :, :], axis=2)

    row_ind, col_ind = linear_sum_assignment(dists)

    matched_distances = dists[row_ind, col_ind]
    valid_matches = matched_distances <= distance_threshold

    tp = int(np.sum(valid_matches))
    fp = int(len(pred) - tp)
    fn = int(len(gt) - tp)

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

    return {
        "tp": tp,
        "fp": fp,
        "fn": fn,
        "precision": float(precision),
        "recall": float(recall),
        "f1": float(f1),
    }


sample_row = mall_df.iloc[0]
result = predict_fidtm_points(
    sample_row["image_path"],
    threshold=FIDTM_THRESHOLD,
    nms_kernel=FIDTM_NMS_KERNEL,
)

loc = compute_localization_metrics(
    gt_points=mall_gt_points_map[sample_row["frame_id"]],
    pred_points=result["points"],
    distance_threshold=16.0,
)

print("Frame:", sample_row["frame_id"])
print("GT count:", sample_row["gt_count"])
print("Pred count:", result["count"])
print(loc)
```

```python
from models.pet import BasePETCount
from models.transformer.utils import window_partition
import torch

def fixed_points_queries_embed_inference(self, samples, stride=8, src=None, **kwargs):
    dense_input_embed = kwargs["dense_input_embed"]
    bs, c = dense_input_embed.shape[:2]

    input_tensor = samples.tensors
    image_shape = torch.tensor(input_tensor.shape[2:])
    shape = (image_shape + stride // 2 - 1) // stride

    shift_x = ((torch.arange(0, shape[1]) + 0.5) * stride).long()
    shift_y = ((torch.arange(0, shape[0]) + 0.5) * stride).long()
    shift_y, shift_x = torch.meshgrid(shift_y, shift_x)

    points_queries = torch.vstack([shift_y.flatten(), shift_x.flatten()]).permute(1, 0)

    h, w = shift_x.shape

    query_embed = dense_input_embed[:, :, points_queries[:, 0], points_queries[:, 1]]
    bs, c = query_embed.shape[:2]

    shift_y_down = points_queries[:, 0] // stride
    shift_x_down = points_queries[:, 1] // stride

    query_feats = src[:, :, shift_y_down, shift_x_down]

    query_embed = query_embed.reshape(bs, c, h, w)
    points_queries = points_queries.reshape(h, w, 2).permute(2, 0, 1).unsqueeze(0)
    query_feats = query_feats.reshape(bs, c, h, w)

    dec_win_w, dec_win_h = kwargs["dec_win_size"]

    query_embed_win = window_partition(
        query_embed,
        window_size_h=dec_win_h,
        window_size_w=dec_win_w,
    )

    points_queries_win = window_partition(
        points_queries,
        window_size_h=dec_win_h,
        window_size_w=dec_win_w,
    )

    query_feats_win = window_partition(
        query_feats,
        window_size_h=dec_win_h,
        window_size_w=dec_win_w,
    )

    div = kwargs["div"]
    div_win = window_partition(
        div.unsqueeze(1),
        window_size_h=dec_win_h,
        window_size_w=dec_win_w,
    )

    valid_div = (div_win > 0.5).sum(dim=0)[:, 0]
    v_idx_gpu = valid_div > 0
    v_idx_cpu = v_idx_gpu.cpu()

    query_embed_win = query_embed_win[:, v_idx_gpu]
    query_feats_win = query_feats_win[:, v_idx_gpu]
    points_queries_win = points_queries_win[:, v_idx_cpu].reshape(-1, 2)

    return query_embed_win, points_queries_win, query_feats_win, v_idx_gpu


BasePETCount.points_queris_embed_inference = fixed_points_queries_embed_inference

print("Correct PET patch installed")
```

```python
@torch.no_grad()
def predict_pet_points(image_path: str, score_threshold: float = 0.5):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(image_path)

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    padded, original_shape = pad_to_pet_size(img)

    original_h, original_w = original_shape
    padded_h, padded_w = padded.shape[:2]

    pil = Image.fromarray(padded)
    tensor = pet_transform(pil).to(DEVICE)
    samples = pet_utils.nested_tensor_from_tensor_list([tensor])

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    start = time.perf_counter()
    output = pet_model(samples, test=True)

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    elapsed = time.perf_counter() - start

    logits = output["pred_logits"][0]
    points = output["pred_points"][0]

    scores = torch.softmax(logits, dim=-1)[:, 1]

    valid = scores >= score_threshold

    points = points[valid].detach().cpu().numpy()
    scores = scores[valid].detach().cpu().numpy()

    pred_points = []

    for p, s in zip(points, scores):
        # PET point format is normalized (y, x)
        y = float(p[0] * padded_h)
        x = float(p[1] * padded_w)

        # Remove points outside original unpadded image
        if 0 <= x < original_w and 0 <= y < original_h:
            pred_points.append((x, y, float(s)))

    return {
        "count": len(pred_points),
        "points": pred_points,
        "time_sec": elapsed,
    }


sample_row = mall_df.iloc[0]

for th in [0.3, 0.4, 0.5, 0.6, 0.7]:
    r = predict_pet_points(sample_row["image_path"], score_threshold=th)
    print(f"threshold={th} | pred_count={r['count']} | gt={sample_row['gt_count']} | time={r['time_sec']:.4f}s")
```

```python
PET_SCORE_THRESHOLD = 0.5

def run_pet_full_metrics(samples_df: pd.DataFrame, distance_threshold: float = 16.0):
    rows = []

    for _, row in tqdm(samples_df.iterrows(), total=len(samples_df)):
        pred = predict_pet_points(
            row["image_path"],
            score_threshold=PET_SCORE_THRESHOLD,
        )

        gt_points = get_gt_points(row["dataset"], row["frame_id"])

        loc = compute_localization_metrics(
            gt_points=gt_points,
            pred_points=pred["points"],
            distance_threshold=distance_threshold,
        )

        pred_count = int(pred["count"])
        gt_count = int(row["gt_count"])

        rows.append({
            "model": "PET",
            "dataset": row["dataset"],
            "frame_id": row["frame_id"],
            "gt_count": gt_count,
            "pred_count": pred_count,
            "abs_error": abs(pred_count - gt_count),
            "sq_error": (pred_count - gt_count) ** 2,
            "time_sec": float(pred["time_sec"]),
            "fps": float(1.0 / pred["time_sec"]) if pred["time_sec"] > 0 else 0.0,
            "tp": loc["tp"],
            "fp": loc["fp"],
            "fn": loc["fn"],
            "precision": loc["precision"],
            "recall": loc["recall"],
            "f1": loc["f1"],
        })

    return pd.DataFrame(rows)


pet_quick_df = run_pet_full_metrics(quick_samples, distance_threshold=16.0)

pet_quick_summary = (
    pet_quick_df
    .groupby(["model", "dataset"])
    .agg(
        frames=("frame_id", "count"),
        mae=("abs_error", "mean"),
        rmse=("sq_error", lambda x: float(np.sqrt(np.mean(x)))),
        avg_fps=("fps", "mean"),
        avg_time_sec=("time_sec", "mean"),
        total_tp=("tp", "sum"),
        total_fp=("fp", "sum"),
        total_fn=("fn", "sum"),
        mean_gt=("gt_count", "mean"),
        mean_pred=("pred_count", "mean"),
    )
    .reset_index()
)

pet_quick_summary["precision"] = (
    pet_quick_summary["total_tp"] /
    (pet_quick_summary["total_tp"] + pet_quick_summary["total_fp"])
)

pet_quick_summary["recall"] = (
    pet_quick_summary["total_tp"] /
    (pet_quick_summary["total_tp"] + pet_quick_summary["total_fn"])
)

pet_quick_summary["f1"] = (
    2 * pet_quick_summary["precision"] * pet_quick_summary["recall"] /
    (pet_quick_summary["precision"] + pet_quick_summary["recall"])
)

display(pet_quick_summary)
```

```python
def extract_points_from_density_map(
    den_map: np.ndarray,
    threshold: float = 0.001,
    nms_kernel: int = 15,
):
    den_map = np.squeeze(den_map)
    den_map = np.maximum(den_map, 0)

    local_max = cv2.dilate(den_map, np.ones((nms_kernel, nms_kernel), np.uint8))
    peaks = (den_map == local_max) & (den_map >= threshold)

    ys, xs = np.where(peaks)

    points = []
    for x, y in zip(xs, ys):
        points.append((float(x), float(y), float(den_map[y, x])))

    return points


@torch.no_grad()
def predict_steerer_points(
    image_path: str,
    threshold: float = 0.001,
    nms_kernel: int = 15,
):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(image_path)

    original_h, original_w = img.shape[:2]

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    x = steerer_transform(img_rgb).unsqueeze(0).to(DEVICE)

    labels = [
        torch.zeros(1, 1024, 2048).to(DEVICE),
        torch.zeros(1, 512, 1024).to(DEVICE),
        torch.zeros(1, 256, 512).to(DEVICE),
        torch.zeros(1, 128, 256).to(DEVICE),
    ]

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    start = time.perf_counter()
    result = steerer_model(x, labels=labels, mode="val")

    if DEVICE == "cuda":
        torch.cuda.synchronize()

    elapsed = time.perf_counter() - start

    den = result["pre_den"]["1"].squeeze().detach().cpu().numpy()

    raw_count = float(den.sum())

    # Extract peaks on resized map 1024x2048
    raw_points = extract_points_from_density_map(
        den,
        threshold=threshold,
        nms_kernel=nms_kernel,
    )

    # Scale points back to original image coordinates
    scale_x = original_w / 2048.0
    scale_y = original_h / 1024.0

    pred_points = []
    for x_map, y_map, score in raw_points:
        x_orig = x_map * scale_x
        y_orig = y_map * scale_y

        if 0 <= x_orig < original_w and 0 <= y_orig < original_h:
            pred_points.append((x_orig, y_orig, score))

    return {
        "count": raw_count,
        "points": pred_points,
        "time_sec": elapsed,
    }


sample_row = mall_df.iloc[0]

for th in [0.0005, 0.001, 0.0015, 0.002, 0.003]:
    r = predict_steerer_points(
        sample_row["image_path"],
        threshold=th,
        nms_kernel=15,
    )
    print(
        f"threshold={th} | raw_count={r['count']:.2f} | "
        f"points={len(r['points'])} | gt={sample_row['gt_count']} | "
        f"time={r['time_sec']:.4f}s"
    )
```
